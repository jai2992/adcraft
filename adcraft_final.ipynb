```python
from diffusers import AutoPipelineForText2Image
import torch

pipe = AutoPipelineForText2Image.from_pretrained("stabilityai/sdxl-turbo", torch_dtype=torch.float32, variant="fp16")
```

#Image model prompting to get image

```python
def generate_image(prompt):
  image = pipe(prompt=prompt, num_inference_steps=1, guidance_scale=0.0).images[0]

  image_bytes = BytesIO()
  image.save(image_bytes, format='PNG')

  image_path = 'generated_img.png'
  image.save(image_path)
```

#Adding brand name and slogen in the generated image

```python
def add_text_to_image(input_image_path, large_text, small_text, output_path='output.png'):
    # Open the image file
    img = Image.open(input_image_path)
    width, height = img.size
    draw = ImageDraw.Draw(img)

    # Specify font size for large and small text
    large_font_size = 50
    small_font_size = 20

    # Specify font .ttf file. Replace 'arial.ttf' with your preferred font.
    font_large = ImageFont.truetype('Anton-Regular.ttf', large_font_size)
    font_small = ImageFont.truetype('DancingScript-SemiBold.ttf', small_font_size)

    # Calculate width and height of the large text
    large_text_width, large_text_height = draw.textsize(large_text, font_large)

    # Calculate the x,y coordinates of the large text
    x_large = width / 2 - large_text_width / 2
    y_large = height / 2 - large_text_height / 2

    # Draw the large text with a black glow
    for adj in range(-3, 4):
        draw.text((x_large+adj, y_large), large_text, font=font_large, fill="black")
        draw.text((x_large, y_large+adj), large_text, font=font_large, fill="black")
    draw.text((x_large, y_large), large_text, font=font_large, fill="white")

    # Calculate width and height of the small text
    small_text_width, small_text_height = draw.textsize(small_text, font_small)

    # Calculate the x,y coordinates of the small text
    x_small = width / 2 - small_text_width / 2
    y_small = y_large + large_text_height + 10  # 10 pixels spacing

    # Draw the small text with a black glow
    for adj in range(-3, 4):
        draw.text((x_small+adj, y_small), small_text, font=font_small, fill="black")
        draw.text((x_small, y_small+adj), small_text, font=font_small, fill="black")
    draw.text((x_small, y_small), small_text, font=font_small, fill="white")

    # Save the image with text
    img.save(output_path)

```

#Getting necessary details from the user to generate the brand poster


```python
category='coffee shop in mountain'
generate_image(category+',  without text , with background blur ')
input_image_path = "generated_img.png"
brand_name = "SPARTA"
slogan = "first sip at the shop."
add_text_to_image(input_image_path, brand_name, slogan)
```
